# Machine Learning Algorithms Using Smote for the imbalance

The pipeline below works in that way that on each fold:
1) It oversamples the minority class

2) Trains the classifier on the remaining folds

3) Validated the classifier on the remaining fold

# Unsupervised Random forest and PCA

```python
print(__doc__)

matplotlib.rcParams['contour.negative_linestyle'] = 'solid'
#Anaomaly Detection using Principle Component Analysis
# Example settings
#Samples to be 399
n_samples = 300
#Outlier Fraction to be used in samples multiplication
outliers_fraction = 0.15
#Defining Outlers
n_outliers = int(outliers_fraction * n_samples)
#Defining Inliers
n_inliers = n_samples - n_outliers

# define outlier/anomaly detection methods to be compared
anomaly_algorithms = [
    ("Robust covariance", EllipticEnvelope(contamination=outliers_fraction)),
    ("One-Class SVM", svm.OneClassSVM(nu=outliers_fraction, kernel="rbf",
                                      gamma=0.1)),
    ("Isolation Forest", IsolationForest(contamination=outliers_fraction,
                                         random_state=42)),
    ("Local Outlier Factor", LocalOutlierFactor(
        n_neighbors=35, contamination=outliers_fraction)),
    #("Unsupervised RF", urf(
    #    n_trees=500)),
    ("Unsupervised RF", urf(
        n_trees=500, max_depth=3))
    ]


# Compare given classifiers under given settings
xx, yy = np.meshgrid(np.linspace(-7, 7, 150),
                     np.linspace(-7, 7, 150))

#Plotting anomaly algorithm defined above
plt.figure(figsize=(len(anomaly_algorithms) * 2 + 3, 12.5))
plt.subplots_adjust(left=.02, right=.98, bottom=.001, top=.96, wspace=.05,
                    hspace=.01)

plot_num = 1
rng = np.random.RandomState(42)

# create a numpy array from the G1 , G2 data, add outliers and fit to the algorithms
# predict with the fitted algorithm(model)
for i_dataset, X in enumerate([portuguese_df[["G1", "G2"]].to_numpy()]):
    # Add outliers
    X = np.concatenate([X, rng.uniform(low=-6, high=6,
                       size=(n_outliers, 2))], axis=0)
#Making unsupervised algorithm from supervised algorithm
    for name, algorithm in anomaly_algorithms:
        t0 = time.time()
        if not 'Unsupervised RF' in name:
            #Fitting algorithm to unsupervised algorithm
            algorithm.fit(X)
        t1 = time.time()
        #PLotting Columns and adding them in unsupervised learning
        plt.subplot(len([portuguese_df[["G1", "G2"]].to_numpy()]), len(anomaly_algorithms), plot_num)
        if i_dataset == 0:
            plt.title(name, size=18)

        # fit the data and tag outliers
        #carry out predictions
        if name == "Local Outlier Factor":
            y_pred = algorithm.fit_predict(X)
            #Making prediction variable only for unsupervised algorithm
        elif 'Unsupervised RF' in name:
            a = algorithm.get_anomaly_score(X, knn=3)
            y_pred = np.ones(a.shape, dtype=int)
            outliers = np.argsort(a)[::-1][:n_outliers]
            y_pred[outliers] = 0
        else:
            y_pred = algorithm.fit(X).predict(X)
            #Making prediction variable for unsupervised algorithm and adding them in unsupervised anamoly detection

        # plot the levels lines and the points
        if (name != "Local Outlier Factor") and (not "Unsupervised RF" in name):  # LOF does not implement predict
            Z = algorithm.predict(np.c_[xx.ravel(), yy.ravel()])
            Z = Z.reshape(xx.shape)
            plt.contour(xx, yy, Z, levels=[0], linewidths=2, colors='black')
            #Making ravel function to fit and transform prediction algorithm

        colors = np.array(['#377eb8', '#ff7f00']) #set up the colors
        plt.scatter(X[:, 0], X[:, 1], s=10, color=colors[(y_pred + 1) // 2])
        # do scatter plot and set both the y and xlimit

        plt.xlim(-7, 7)
        plt.ylim(-7, 7)
        plt.xticks(())
        plt.yticks(())
        plt.text(.99, .01, ('%.2fs' % (t1 - t0)).lstrip('0'),
                 transform=plt.gca().transAxes, size=15,
                 horizontalalignment='right')
        plot_num += 1
        #Plotting PCA function

plt.show()
```

```python
# set the principal components to a dataframe
G_data = pd.DataFrame(data = G_components
             , columns = ['principal component 1', 'principal component 2', 'principal component 3'])
             #Making PCA columns 
```

```python
# join with the other data
G_df = pd.concat([G_data, portuguese_df[['G3']]], axis = 1)
#Making a new dataframe and with target variable
```

```python
# plot the principal components separately to get a view of their boundaries
fig = plt.figure(figsize = (8,8))
ax = fig.add_subplot(1,1,1) 
ax.set_xlabel('Principal Component 1', fontsize = 15)
ax.set_ylabel('Principal Component 2', fontsize = 15)
ax.set_title('3 component PCA', fontsize = 20)
targets = G_df["G3"].unique()
colors = ['r', 'g', 'b']
for target, color in zip(targets,colors):
    indicesToKeep = G_df["G3"] == target
    ax.scatter(G_df.loc[indicesToKeep, 'principal component 1']
               , G_df.loc[indicesToKeep, 'principal component 2']
               , c = color
               , s = 50)
ax.legend(targets)
ax.grid()
plt.savefig('PCA.png')
#Plotting PCA 
```
